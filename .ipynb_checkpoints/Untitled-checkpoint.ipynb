### Lesson No. 1

#### 1. Why are successful cities dense, expensive, and with few large green patches?

Successful cities are dense because of their growth in population. Successful cities attract companies and private households alike. This also results in rents becoming more expensive. Too many green patches are not efficient because they take too much space, still, cities need parks for keeping their attractiveness. 
A few remarkable, large green patches are not only recreation areas for citizens but can also become famous sights of the cities.

#### 2. Why do cities need policy interventions?

Interventions that classified in taxation, regulation, investment, information are necessary if a city wants to lead the city's developement in a specific direction. 
Some examples where cities should intervene are when less competetive land uses are driven out of a dense area because they are not seen as economically important or if dense areas produce clusters with side effects that can harm the developement of the city e.g. air pollution or crime or if there is not enough housing for the growing population of the city. Cities are also competing against each other and policy interventions that attract industries are also used.

### Lesson No. 5

#### 8. How do cellular automata see and explain the world?

The basic cellular automata does only see cells, states, neighborhoods and transition rules.
Arranged in a n*k lattice, the cells are can have different states. If a neighbouring cell has a different state, the state of the cell is changed in the next step according to the transition rules. In urban planning cells are representing landspace and the two states are developed (built-in) and undeveloped (natural) land. Neighbourhoods are defined by the von-Neumann or Moore contiguity or a modification of those.

Cellular automatas try to replicate the growth process of a city. If the CA can replicate the history of growth correctly, it is most likely that it can predict the future development, too. For a precise model the CA has to be adjusted because the environment is more complex. One approach is the SLEUTH-model.
